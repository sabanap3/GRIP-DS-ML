#### From this graph, one can easily makeout that which category & sub-category choose, when they are looking to purchase a product. 

```python
plt.figure(figsize=(12,10))
df['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

#### The store has wide variety of Office Supplies especially in Binders and Paper department. 

```python
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total profit ans sales per Sub-Category')
#plt.legend('Profit')
#plt.legend('Sales')
plt.show()
```

#### Highest profit is earned in Copiers while selling price for chairs and phones is extremely high compared to other products.
#### Another interesting fact-people don't prefer to buy tables and bookcases from superstore.Hence these departments are in loss.

```python
#Count os Sub-Category region wise
plt.figure(figsize=(15,5))
sns.countplot(x='Sub-Category',hue='Region',data=df)
plt.show()
```

#### People residing in western part of US tend to order more from superstore.

```python
df['Cost']=df['Sales']-df['Profit']
df['Cost'].head()
```

```python
df['Profit %']=(df['Profit']/df['Cost'])*100
```

```python
#profit percentage of first five product
df.iloc[[0,1,2,3,4],[7,13]]
```

```python
#products with high profit percentage
df.sort_values(['Profit %','Category'],ascending=False).groupby('Profit %').head(5)
```

#### Retailers selling Machines,Phones,Copiers have got 100% Profit in their business.

```python
fig=plt.figure(figsize=(5,5))
ax=fig.add_subplot(111)
s=sns.countplot('Segment',data=df)
for s in ax.patches:
    ax.annotate('{:.0f}'.format(s.get_height()),(s.get_x()+0.15,s.get_height()+1))
plt.show()
```

#### This distribution is heighest in consumer segment.

```python
#Calculateing profit gained in each Category
fig=plt.figure(figsize=(16,8))
ax=fig.add_subplot(111)
sns.barplot('Ship Mode','Profit %',hue='Sub-Category',palette='Paired',data=df)
for o in ax.patches:
    ax.annotate('{:.0f}'.format(o.get_height()),(o.get_x()+0.15,o.get_height()+1))
plt.show()
```

```python
#sales per year
df.groupby('Ship Mode')['Sales','Profit %'].agg(['sum']).plot.bar()
plt.title('Ship mode wise total sales and % of profit gained')
```

```python
profit_city = df.groupby("City").sum()['Profit']
```

```python
profit_city.sort_values(ascending = False).head()
```

Cities which have the maximum loss

```python
#plotting
fig,axes = plt.subplots(1,2, figsize = (15,5))
profit_city.sort_values(ascending = False).head().plot(kind ='bar', ax = axes[0])
axes[0].tick_params(labelrotation = 10)
axes[0].set_title("MAXIMUM PROFIT OF CITY")

profit_city.sort_values(ascending = True).head().plot(kind='bar',ax = axes[1])
plt.xticks(rotation = 15)
plt.title("MINIMUM PROFIT OF CITY ")

plt.show()
```

New York City has maximum profit and Philadelphia has maximum loss 

```python
df3=df.groupby('Region').sum()[['Profit','Sales']]
df3['GrossProfit']= (df3['Profit']/df3['Sales'])*100
plt.bar(df3.index.values,df3['GrossProfit'],color=['Red','Blue','Green','Orange'])
plt.xlabel('Region')
plt.title('Grossprofit margin for region')
print(df3)
```
